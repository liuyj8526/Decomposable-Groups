```python
n = input().split()
```

```python
a = ["1"]
b = [] 
print(n)
while len(n) != 0:
    k = n.pop(0)
    if a[-1] != k:
        a.append(k)
        te = ""
        te += n.pop(0) + "  "
        b.append(te)
    else: 
        b[-1] = b[-1] + n.pop(0) + "  "
for a1 in a:
    print(a1)
for b1 in b:
    print(b1)
```
