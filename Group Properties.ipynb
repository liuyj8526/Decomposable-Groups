```python
"""Testing properties of groups"""
class testGrp():
    def subgroups(self, G):
        subgroups = G.conjugacy_classes_subgroups()
#         print(subgroups)
        return subgroups

    """Test if a subgroup is frobenius. Reference: https://www.mail-archive.com/forum@mail.gap-system.org/msg04060.html"""
    def testFrob(self, G):
        if self.testCenter(G) == True:
            return False
        fit = G.fitting_subgroup()
        if fit.order() == 1 or fit.order() == G.order():
            return False
        reps = gap.ComplementClassesRepresentatives(G,fit)
        if len(reps) != 1:
            return False
        test_elem = fit.list()
        elem = G.list()
        for rep in reps:
            frob = True
            perm_group = PermutationGroup(gap_group = rep.AsPermGroup())
            target = perm_group.order()
            elem_gen = fit.list()
            elem_gen += perm_group.list()
            for g in test_elem[1:]:
                conj = perm_group.conjugate(g)
                intt = perm_group.intersection(conj)
                if (intt.order() == 1 or intt.order() == target) == False:
                    frob = False
                elem_gen += conj.list()
            for g in elem:
                if g not in elem_gen:
                    frob = False
            if frob == True:
                return True
        return False
    
    """Test if a group as nontrivial center"""    
    def testCenter(self, G):
        print(str(gap.IdSmallGroup(G.center())[1]) + "," + str(gap.IdSmallGroup(G.center())[2]))
#         if G.center().order() >1:
#             return True
        return False

    """For finding hp group in original problem. p is the prime"""
    def hp_group(self, G, p):
        elems = G.list()
        gens = []
        for g in elems:
            if g.order() != p:
                gens.append(g)
        hp_group = PermutationGroup(gens)
        return hp_group
    
    """For finding hp group in general case. i is the index of the subgroup, and p is the prime"""
    def hp_group_gen(self, G, i, p):
        i = 0
        for g in G.list():
            if g.order() == 2:
                print(g)
                i += 1
        print(i)
        b = self.subgroups(G)
        H1 = b[i]
        elems = G.list()
        gens = []
        gens += H1.list()
        X = []
        for g in elems:
            for h in H1.conjugate(g).list():
                if h not in X:
                    X.append(h)
        for g in elems:
            if g not in X and g.order() != p:
                gens.append(g)
        print(len(gens))
        hp_group = PermutationGroup(gens)
        return hp_group
```
